```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/auto_insurance_sweden.csv")
plt.plot(df["claims"], df["payment"], "*")
plt.xlabel("Claims")
plt.ylabel("Payment");
```

## USING ALGEBRA

```python
x,y=df["claims"],df["payment"]
covar=np.cov(x,y)[0][1]
var_x=np.cov(x,y)[0][0]
m=covar/var_x
n=y.mean()-m*x.mean()
print("Slope: ",m)
print("Intercept: ",n)
xs=[-5,130]
ys=[n,130*m+n]
plt.plot(x,y,"*")
plt.plot(xs,ys,color="orange")
plt.xlabel("Claims")
plt.ylabel("Payment");
```

## USING GRADIENT DESCENT

```python
X,Y=x,y
N=len(x)
step=0.0001
m,n=0,0
plt.figure(1)
plt.plot(X,Y,"*")
plt.plot([0,120],[n,120*m+n])
ind=1
for i in range(50000):
    Y_pred=m*X+n
    Dm=-2*sum(X*(Y-Y_pred))/N
    Dn=-2*sum((Y-Y_pred))/N
    m-=step*Dm
    n-=step*Dn
    if i in [1,2,5,15]:
        plt.figure(ind+1)
        ind+=1
        plt.plot(X,Y,"*")
        plt.plot([0,120],[n,120*m+n])
plt.figure(ind+1)
plt.plot(X,Y,"*")
plt.plot([0,120],[n,120*m+n])

print("Slope: ",m)
print("Intercept: ",n)
```
